# Red Wine Quality Investigation!

Wine has been a popular beverage of humankind for thousands of years. The earliest wine ever known was from 8000-6000 BC.

So basically, Red Wine is an alcoholic beverage made by fermenting the juice of dark-skinned grapes.
Red Wine differs from White Wine (Regular Wine) in its base material and production process.
Red Wine is made with dark-skinned rather than light-skinned grapes. During red wine production,
the winemaker allows pressed grape juice, called must, to macerate and ferment with the dark grape skins,
which adds colour, flavour and tannin to the wine. Alcohol occurs when yeast converts grape sugar into
Ethanol and Carbon Dioxide.

Sometimes people would like to know the quality of Red Wine because there are a lot of brands out there. But they don’t know how to analyze it, they don’t know what kind of specific conditions which is affecting the quality of the Red Wine apart from the years.

To solve this problem, we can use Python to analyze available data. These data can be used to predict the quality of the Red Wine given certain variables or factors.

To define the quality of Red Wine, there are a lot of factors. It could be the Alcohol (ABV), Sulphates, pH (a measure of how acidic/basic liquid is), etc.
Dataset

The dataset used in this article is based on the ingredients or the composition of mostly Red Wine. This means it has to include:

<ol>
<li>Fixed Acidity: Most acids involved with wine or fixed or non-volatile (do not evaporate readily)</li>
<li>Volatile Acidity: The amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste</li>
<li>Citric Acid: Often added to wines to increase acidity, complement a specific flavor or prevent ferric hazes</li>
<li>Residual Sugar: From the natural grape sugars left in a wine after the alcoholic fermentation finishes.</li>
<li>Chlorides: The amount of salt in the wine</li>
<li>Free Sulfur Dioxide: It prevents microbial growth and the oxidation of wine</li>
<li>Total Sulfur Dioxide: The amount of free + bound forms of SO₂</li>
<li>Density: Sweeter wines have a higher density</li>
<li>pH: Describes the level of acidity on a scale of 0–14. Most wines are always between 3–4 on the pH scale</li>
<li>Alcohol: Available in small quantities in wines makes the drinkers sociable</li>
<li>Sulphates: A wine additive that contributes to SO₂ levels and acts as an antimicrobial and antioxidant</li>
<li>Quality: which is the output variable/predictor</li>

```python
# Define the boundaries for the ranks
bins = [0, 0.4, 0.8, np.inf]
names = ['0', '1', '2']

# Create the 'volatile acidity rank' column
df['volatile acidity rank'] = pd.cut(df['volatile acidity'], bins, labels=names)
```

```python
df.head()
```

```python
#Calculate mean fixed acidity & citric acid for each volatile acidity rank
#used groupby
mean_values = df.groupby('volatile acidity rank')[['fixed acidity', 'citric acid']].mean()

print(mean_values)
```

```python
#to show subset of rows in which chlorides >= 0.1 and residual sugar == 1.6
#subsetdataframe = dataframe[dataframe['series'] >= 0.1]
subset_df = df[(df['chlorides'] >= 0.1) & (df['residual sugar'] == 1.6)]
print(subset_df)

```

```python
#to plot a histogram with 10 bins. Figure size is 20, 20

fig, axes = plt.subplots(5, 5, figsize=(20, 20))
columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
           'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
           'pH', 'sulphates', 'alcohol', 'quality']

# Flatten the axes array and zip with columns
for ax, col in zip(axes.flatten(), columns):
    ax.hist(df[col], bins=10, edgecolor='black')
    ax.set_title(f'Histogram of {col}')

# Remove unused subplots
for ax in axes.flatten()[len(columns):]:
    ax.remove()

plt.tight_layout()
plt.show()
```

```python
# to create a heatmap using seaborn library
#import seaborn library
#correlation matrix

import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr()

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True,ax=ax)
plt.show()
```

```python
#to plot a scatter plot for volatile acidity and quality

import matplotlib.pyplot as plt

plt.scatter(df['volatile acidity'], df['quality'])
plt.title('Scatter plot of volatile acidity vs quality')
plt.xlabel('volatile acidity')
plt.ylabel('quality')
plt.show()
```

# **Finding:**




Based on the above heat map, the wine quality is highly corelated to volatile acidity and alcohol. There is a strong positive relationship between alcohol and quality of wine, which means higher alcohol percentage is associated with higher quality of the red wine. Furthermore, there is a strong negative correlation between volatile acidity and quality of wine, which means higher volatile acidity is associated with lower quality of the red wine. Also, pH and citric acid & fixed acidity are highly inversely related due to acids have smaller pH values. The alcohol percentage in red wines also has a negative correlation with density, which means when alcohol percentage increases, its density decreases.


Besides, when calculating mean fixed acidity and citric acid for each volatile acidity rank using group by, I observe that those red wines with higher volatile acidity has lesser amount of citric acid and fixed acidity.

Therefore, based on the data visualization, a higher quality of red wines has higher level of alcohol with lower density and lower volatile acidity. However, there may be some other variables such as year the wine was produced may be influence the quality of red wines.
